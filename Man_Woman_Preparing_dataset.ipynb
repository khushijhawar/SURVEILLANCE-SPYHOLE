```
#creating a folder for test data
os.makedirs("temp_test_data")
all_man = os.listdir("dataset/test/man")
all_woman = os.listdir("dataset/test/woman")
m = len(all_man)
w = len(all_woman)
ind = 0
for ind in range(len(all_man)):
  source = "dataset/test/man/"+all_man[ind]
  destination = f"temp_test_data/img{ind}.jpg"
  shutil.move(source,destination)


for i in range(len(all_woman)):
  source = "dataset/test/woman/"+all_woman[i]
  destination = f"temp_test_data/img{ind}.jpg"
  shutil.move(source,destination)
  ind+=1
```

```
#mixing up the images and finalising the test data
import random
os.makedirs("test_data")
images = os.listdir("temp_test_data")
ind = 0
while(len(images)!=0):
  image = random.choice(images)
  source = "temp_test_data/"+image
  destination = f"test_data/img{ind}.jpg"
  shutil.move(source,destination)
  ind+=1
  images.remove(image)
```

```
#changing size of image data

from PIL import Image
import os, sys

path = "test_data"
dirs = os.listdir( path )

os.makedirs("Dataset/test_data")
def resize():
  i=0
  for item in dirs:
    im = Image.open(path+ "/"+item)
    imResize = im.resize((200,200), Image.ANTIALIAS)
    imResize.save(f'Dataset/test_data/resized{i}.jpg', 'JPEG', quality=90)
    i+=1

resize()
```
